### Query Params Description

### Benzinga API Pagination Parameters

**Page & PageSize Parameters**

`page`:
- Skip counter (starts at 0)
- Range: 0-100000 
- Formula: Starting Item = page * pageSize

`pageSize`:
- Items per request
- Range: 1-100 (default: 15)

**Example** (300 total items):

`pageSize=100`:

page 0 → items 1-100                
page 1 → items 101-200              
page 2 → items 201-300              


`pageSize=50`:

page 0 → items 1-50
page 1 → items 51-100
page 2 → items 101-150
page 3 → items 151-200
page 4 → items 201-250
page 5 → items 251-300

### Benzinga API Query Parameters

`displayOutput`:
- Controls article content length
- Values: 
  - `headline` (default): title only
  - `abstract`: title + teaser
  - `full`: complete article content

`date`/`dateFrom`/`dateTo`:
- `date`: Single date query (shorthand for identical dateFrom/dateTo)
- `dateFrom`: Start date for range
- `dateTo`: End date for range
- Format: Standard date string

`updatedSince`/`publishedSince`:
- `updatedSince`: Last updated Unix timestamp (UTC)
- `publishedSince`: Last published Unix timestamp (UTC)
- Used for delta updates/real-time tracking

`sort`:
- Controls result ordering
- Format: `field:direction`
- Fields: `id`, `created`, `updated`
- Directions: `asc`, `desc`
- Examples: 
  - `id:asc`
  - `created:desc`
  - `updated:asc`

`isin`/`cusip`:
- `isin`: International Securities Identification Numbers
- `cusip`: Committee on Uniform Security Identification Procedures
- Limit: Maximum 50 identifiers, comma-separated
- Note: CUSIP requires license agreement

`tickers`:
- Stock symbols
- Limit: Maximum 50 tickers
- Format: Comma-separated (e.g., "AAPL,MSFT,GOOGL")

`channels`:
- News categories
- Format: Comma-separated channel names/IDs
- Example: "WIIM" for "Why is it Moving" data

`topics`:
- Searches Title, Tags, and Body
- Format: Comma-separated words/phrases
- Prioritizes matches in order: Title > Tags > Body

`authors`:
- Filter by content creators
- Format: Comma-separated author names

`content_types`:
- Filter by type of content
- Format: Comma-separated content types

```python
import requests
import schedule
import time
import json
from datetime import datetime
from typing import Dict, List, Any
from email.utils import parsedate_to_datetime




# API Configuration
API_URL = "https://api.benzinga.com/api/v2/news"
API_HEADERS = {"accept": "application/json"}
API_PARAMS = {
    "token": BENZINGANEWS_API_KEY,
    "page": 0,
    "pageSize": 10,
    "displayOutput": "full"
}

def extract_list_field(item: Dict[str, Any], field: str) -> str:
    """Extract and join list fields like stocks, channels, tags"""
    elements = [element['name'] for element in item.get(field, [])]
    return ', '.join(elements) if elements else 'None'

def process_news_item(item: Dict[str, Any]) -> Dict[str, str]:
    """Process a single news item and return formatted fields"""
    return {
        'id': str(item.get('id', 'No ID')),
        'title': item.get('title', 'No Title'),
        'created': item.get('created', 'Unknown Date'),
        'updated': item.get('updated', 'Unknown Date'),
        'url': item.get('url', 'No URL'),
        'author': item.get('author', 'Unknown'),
        'stocks': extract_list_field(item, 'stocks'),
        'channels': extract_list_field(item, 'channels'),
        'tags': extract_list_field(item, 'tags')
    }



def parse_date(date_str: str) -> datetime:
    try:
        # Parse RFC 2822 formatted string
        return parsedate_to_datetime(date_str)
    except Exception:
        return None  # Handle invalid date format

def print_news_item(news_item: Dict[str, str]) -> None:
    """Print formatted news item"""

    # Inline function to parse date strings
    def parse_date(date_str: str) -> datetime:
        try:
            # Parse RFC 2822 formatted string
            return parsedate_to_datetime(date_str)
        except Exception:
            return None  # Handle invalid date format

    print(f"ID: {news_item['id']}")
    print(f"Title: {news_item['title']}")
    print(f"Created: {news_item['created']}")
    print(f"Updated: {news_item['updated']}")

    # Parse the 'updated' field
    updated_item = parse_date(news_item['updated'])
    if updated_item:
        print(f"Updated Type: {type(updated_item)}")
        print(f"Updated Date: {updated_item.date()}")
        print(f"Updated Time: {updated_item.time()}")
        print(f"Updated Timezone: {updated_item.tzinfo}")
        print(f"Updated Timestamp: {updated_item.timestamp()}")
    else:
        print("Updated field is not a valid datetime string.")

    print(f"URL: {news_item['url']}")
    print(f"Author: {news_item['author']}")
    print(f"Stocks: {news_item['stocks']}")
    print(f"Channels: {news_item['channels']}")
    print(f"Tags: {news_item['tags']}")
    print("-" * 80 + "\n")



# def print_news_item(news_item: Dict[str, str]) -> None:
#     """Print formatted news item"""
#     print(f"ID: {news_item['id']}")
#     print(f"Title: {news_item['title']}")
#     print(f"Created: {news_item['created']}")
#     print(f"Updated: {news_item['updated']}")    

#     print(f"URL: {news_item['url']}")
#     print(f"Author: {news_item['author']}")
#     print(f"Stocks: {news_item['stocks']}")
#     print(f"Channels: {news_item['channels']}")
#     print(f"Tags: {news_item['tags']}")
#     print("-" * 80 + "\n")

def fetch_news() -> List[Dict[str, Any]]:
    """Fetch news from Benzinga API"""
    try:
        response = requests.get(API_URL, headers=API_HEADERS, params=API_PARAMS)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"API Request Error: {e}")
        if hasattr(e, 'response'):
            print(f"Response Status: {e.response.status_code}")
            print(f"Response Text: {e.response.text[:200]}...")  # First 200 chars
    except json.JSONDecodeError as e:
        print(f"JSON Parsing Error: {e}")
    except Exception as e:
        print(f"Unexpected Error: {e}")
    return []

def fetch_and_process_news() -> None:
    """Main function to fetch and process news"""
    news_items = fetch_news()
    if news_items:
        print(f"\nFetched {len(news_items)} news items at {datetime.now()}\n")
        for item in news_items:
            processed_item = process_news_item(item)
            print_news_item(processed_item)
    else:
        print("No news items retrieved")

def main():
    """Main entry point"""
    print("Starting Benzinga News Fetcher...")
    
    # Initial run
    fetch_and_process_news()
    
    # Schedule subsequent runs
    schedule.every(5).seconds.do(fetch_and_process_news)
    
    # Main loop
    while True:
        try:
            schedule.run_pending()
            time.sleep(5)
        except KeyboardInterrupt:
            print("\nStopping news fetcher...")
            break
        except Exception as e:
            print(f"Main loop error: {e}")
            time.sleep(5)

if __name__ == "__main__":
    main()
```

```python
import requests
import schedule
import time
from datetime import datetime, timedelta, timezone as dt_timezone

from email.utils import parsedate_to_datetime
from typing import Dict, List, Any
from pytz import timezone as pytz_timezone



# API Key and Configuration
API_URL = "https://api.benzinga.com/api/v2/news"
API_HEADERS = {"accept": "application/json"}
API_PARAMS = {
    "token": BENZINGANEWS_API_KEY,
    "page": 0,
    "pageSize": 10,
    "displayOutput": "full"
}

# Initialize latest_updated to 1 hour ago
latest_updated = int((datetime.now(dt_timezone.utc) - timedelta(hours=1)).timestamp())



def to_local_time(utc_time, tz_name="US/Eastern"):
    """Convert UTC datetime to the specified timezone (e.g., EST/EDT)."""
    local_tz = pytz_timezone(tz_name)
    return utc_time.astimezone(local_tz)



def rfc2822_to_unix(date_str: str) -> int:
    """Convert RFC 2822 formatted date (GMT-4:00) to Unix timestamp in UTC."""
    try:
        dt = parsedate_to_datetime(date_str)
        # Force the API's GMT-4:00 offset to UTC        
        return int(dt.astimezone(dt_timezone.utc).timestamp())

    except Exception as e:
        print(f"Error parsing date: {e}")
        return 0


def fetch_news_with_delta(latest_updated: int) -> List[Dict[str, Any]]:
    """Fetch news with updatedSince parameter to minimize data retrieval."""
    params = API_PARAMS.copy()
    if latest_updated:
        params["updatedSince"] = latest_updated - 5  # Subtract 5 seconds as per API guidelines

    try:
        response = requests.get(API_URL, headers=API_HEADERS, params=params)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"API Request Error: {e}")
    return []

def process_news_item(item: Dict[str, Any]) -> Dict[str, str]:
    """Process a single news item and return formatted fields."""
    return {
        'id': str(item.get('id', 'No ID')),
        'title': item.get('title', 'No Title'),
        'created': item.get('created', 'Unknown Date'),
        'updated': item.get('updated', 'Unknown Date'),
        'url': item.get('url', 'No URL'),
        'author': item.get('author', 'Unknown'),
        'stocks': ', '.join([stock['name'] for stock in item.get('stocks', [])]),
        'channels': ', '.join([channel['name'] for channel in item.get('channels', [])]),
        'tags': ', '.join([tag['name'] for tag in item.get('tags', [])])
    }

def print_news_item(news_item: Dict[str, str]) -> None:
    """Print formatted news item."""
    try:
        updated_time = parsedate_to_datetime(news_item['updated']).astimezone(dt_timezone.utc)
        local_time = to_local_time(updated_time)
    except Exception as e:
        updated_time, local_time = "Invalid", "Invalid"
        print(f"Error parsing 'updated': {e}")

    print(f"ID: {news_item['id']}")
    print(f"Title: {news_item['title']}")
    print(f"Created: {news_item['created']}")
    print(f"Updated (UTC): {updated_time}")
    print(f"Updated (Local): {local_time}")
    print(f"URL: {news_item['url']}")
    print(f"Author: {news_item['author']}")
    print(f"Stocks: {news_item['stocks']}")
    print(f"Channels: {news_item['channels']}")
    print(f"Tags: {news_item['tags']}")
    print("-" * 80)



def fetch_and_process_news() -> None:
    """Main function to fetch and process news."""
    global latest_updated
    news_items = fetch_news_with_delta(latest_updated)

    if news_items:
        print(f"\nFetched {len(news_items)} news items at {datetime.now()}")
        for item in news_items:
            processed_item = process_news_item(item)
            print_news_item(processed_item)

            # Update the latest_updated timestamp
            updated_unix = rfc2822_to_unix(item['updated'])
            if not latest_updated or updated_unix > latest_updated:
                latest_updated = updated_unix
    else:
        print("No new or updated news items retrieved")


def main():
    """Main entry point."""
    print("Starting Benzinga News Fetcher...")

    # Initial run
    fetch_and_process_news()

    # Schedule subsequent runs
    schedule.every(5).seconds.do(fetch_and_process_news)

    # Main loop
    while True:
        try:
            schedule.run_pending()
            time.sleep(5)
        except KeyboardInterrupt:
            print("\nStopping news fetcher...")
            break

if __name__ == "__main__":
    main()
```

### Polling Optimization Checklist with Documentation References

  * Doc Reference: "set updatedSince to a value 5 seconds earlier than you actually want"
  * Critical Note: Required to handle latency and second-only timestamps

**Timestamp Handling**
- [x] Convert EDT timezone to Unix timestamp
  * Doc Reference: Time zone is "Greenwich Mean Time (GMT) - 4:00, commonly known as Eastern Daylight Time (EDT)"
- [x] Track max 'updated' time from current batch
  * Doc Reference: Using "maximum value of the updated field (converted to a Unix timestamp) in rows already pulled"

**Pagination Implementation**
- [x] Continue fetching pages until one of:
  * Received fewer items than pageSize
  * No more items returned
  * Error occurred

```python
import requests
import schedule
import time
import json
from datetime import datetime, timezone, timedelta
from typing import Dict, List, Any, Optional
import pytz

MAX_PAGE_LIMIT = 100000  # Added safety

# Global timestamp tracker
LATEST_UPDATE_TIMESTAMP: Optional[int] = None

# API Configuration
API_URL = "https://api.benzinga.com/api/v2/news"
API_HEADERS = {"accept": "application/json"}
API_PARAMS = {
    "token": BENZINGANEWS_API_KEY,
    "page": 0,
    "pageSize": 100,  # Increased for efficiency
    "displayOutput": "full",
    "sort": "updated:desc"  # Sort by most recent updates
}


def format_to_edt(dt: datetime) -> str:
    """Format datetime to EDT string - Benzinga already provides EDT times"""
    edt_tz = pytz.timezone('America/New_York')
    dt = dt.astimezone(edt_tz)  # Just ensure it's in EDT
    return dt.strftime('%a, %d %b %Y %H:%M:%S -0400')



def get_initial_timestamp() -> int:
    """Get Unix timestamp from one hour ago in EDT"""
    edt_tz = pytz.timezone('America/New_York')
    now = datetime.now(edt_tz)
    one_hour_ago = now - timedelta(hours=1)
    return int(one_hour_ago.timestamp()) - 5

def convert_edt_to_unix(edt_time_str: str) -> int:
    """Convert Benzinga's EDT timestamp string to Unix timestamp"""
    try:
        # Parse directly as EDT time (-0400)
        dt = datetime.strptime(edt_time_str, '%a, %d %b %Y %H:%M:%S -0400')
        edt_tz = pytz.timezone('America/New_York')
        dt = edt_tz.localize(dt)
        return int(dt.timestamp())
    except Exception as e:
        print(f"Time conversion error: {e}")
        return 0



def extract_list_field(item: Dict[str, Any], field: str) -> str:
    """Extract and join list fields like stocks, channels, tags"""
    elements = [element['name'] for element in item.get(field, [])]
    return ', '.join(elements) if elements else 'None'

def process_news_item(item: Dict[str, Any]) -> Dict[str, str]:
    """Process a news item from Benzinga API"""
    try:
        return {
            'id': str(item.get('id', 'No ID')),
            'title': item.get('title', 'No Title'),
            'created': item.get('created', ''),  # Use original EDT string
            'updated': item.get('updated', ''),  # Use original EDT string
            'url': item.get('url', 'No URL'),
            'author': item.get('author', 'Unknown'),
            'stocks': extract_list_field(item, 'stocks'),
            'channels': extract_list_field(item, 'channels'),
            'tags': extract_list_field(item, 'tags')
        }
    except ValueError as e:
        print(f"Error processing news item: {e}")
        return {}

def print_news_item(news_item: Dict[str, str]) -> None:
    """Print formatted news item"""
    print(f"ID: {news_item['id']}")
    print(f"Title: {news_item['title']}")
    print(f"Created: {news_item['created']}")
    print(f"Updated: {news_item['updated']}")
    print(f"URL: {news_item['url']}")
    print(f"Author: {news_item['author']}")
    print(f"Stocks: {news_item['stocks']}")
    print(f"Channels: {news_item['channels']}")
    print(f"Tags: {news_item['tags']}")
    print("-" * 80 + "\n")


def get_lagged_timestamp(max_updated_time: str) -> int:
    """
    Get Unix timestamp (UTC) from 5 seconds before the given EDT time.
    Input: Benzinga EDT timestamp string
    Output: Unix timestamp (UTC) with 5-second buffer
    """
    unix_time = convert_edt_to_unix(max_updated_time)
    return unix_time - 5


def fetch_news() -> List[Dict[str, Any]]:
    """Fetch news from Benzinga API with efficient polling"""
    global LATEST_UPDATE_TIMESTAMP
    all_items = []
    current_page = 0
    
    while current_page < MAX_PAGE_LIMIT:
        try:
            current_params = API_PARAMS.copy()
            current_params['page'] = current_page

            if LATEST_UPDATE_TIMESTAMP:
                print("\nTimestamp Debug:")
                print(f"LATEST_UPDATE_TIMESTAMP (Unix): {LATEST_UPDATE_TIMESTAMP}")
                edt_tz = pytz.timezone('America/New_York')
                print(f"LATEST_UPDATE_TIMESTAMP (EDT): {format_to_edt(datetime.fromtimestamp(LATEST_UPDATE_TIMESTAMP, edt_tz))}")
                
                # API docs: "set updatedSince to a value 5 seconds earlier"
                current_params['updatedSince'] = LATEST_UPDATE_TIMESTAMP - 5

            response = requests.get(API_URL, headers=API_HEADERS, params=current_params)
            response.raise_for_status()
            news_items = response.json()

            if news_items:
                first_item = news_items[0]
                print("\nFirst Item Timestamps:")
                print(f"Created: {first_item.get('created')}")
                print(f"Updated: {first_item.get('updated')}")
                print(f"Our current time (EDT): {format_to_edt(datetime.now(pytz.timezone('America/New_York')))}")

            if not news_items:
                break
            
            all_items.extend(news_items)
            
            if len(news_items) < API_PARAMS['pageSize']:
                break
                
            current_page += 1
            
        except Exception as e:
            print(f"Error in fetch_news: {str(e)}")
            return all_items
    
    # Update timestamp from most recent item
    if all_items:
        max_updated_time = max(item.get('updated', '') for item in all_items)
        if max_updated_time:
            LATEST_UPDATE_TIMESTAMP = convert_edt_to_unix(max_updated_time)
            print(f"Updated timestamp to: {LATEST_UPDATE_TIMESTAMP} "
                  f"(EDT: {format_to_edt(datetime.fromtimestamp(LATEST_UPDATE_TIMESTAMP, timezone.utc))} "
                  f"from max time: {max_updated_time}")
    
    return all_items


def fetch_and_process_news() -> None:
    """Main function to fetch and process news"""
    global LATEST_UPDATE_TIMESTAMP
    
    if LATEST_UPDATE_TIMESTAMP is None:
        LATEST_UPDATE_TIMESTAMP = get_initial_timestamp()
        edt_time = format_to_edt(datetime.fromtimestamp(LATEST_UPDATE_TIMESTAMP, timezone.utc))
        print(f"Initialized timestamp: {LATEST_UPDATE_TIMESTAMP} (EDT: {edt_time})")

    
    news_items = fetch_news()
    if news_items:
        current_time_edt = format_to_edt(datetime.now(pytz.timezone('America/New_York')))
        print(f"\nFetched {len(news_items)} news items at {current_time_edt}\n")
        for item in news_items:
            processed_item = process_news_item(item)
            print_news_item(processed_item)
    else:
        current_time_edt = format_to_edt(datetime.now(pytz.timezone('America/New_York')))
        print(f"No news items retrieved at {current_time_edt}")


def main():
    """Main entry point"""
    print("Starting Benzinga News Fetcher...")
    
    # Initial run
    fetch_and_process_news()
    
    # Schedule subsequent runs
    schedule.every(5).seconds.do(fetch_and_process_news)
    
    # Main loop
    while True:
        try:
            schedule.run_pending()
            time.sleep(5)
        except KeyboardInterrupt:
            print("\nStopping news fetcher...")
            break
        except Exception as e:
            print(f"Main loop error: {e}")
            time.sleep(5)

if __name__ == "__main__":
    main()
```
